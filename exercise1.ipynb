## 1.1 The Two-Body-Problem

The two body problem can be written as

$$
\begin{align*}
\vec{F}_1 &= m_1 \ddot{\vec{r}}_1 = \frac{G m_1 m_2}{|\vec{r}_1 - \vec{r}_2|^3} \cdot (\vec{r}_2 - \vec{r}_1)\\
\vec{F}_2 &= m_2 \ddot{\vec{r}}_2 = \frac{G m_1 m_2}{|\vec{r}_1 - \vec{r}_2|^3} \cdot (\vec{r}_1 - \vec{r}_2)
\end{align*}
$$

where Newton’s second law applies to objects 1 and 2, with $F_1$ being the force exerted on object 1 and $F_2$ the force exerted on object 2.

**a) Write the system out in its components.**

$$
\begin{align*}
m_1 \cdot \ddot{x}_1 &= \frac{G \cdot m_1 \cdot m_2}{\left(\sqrt{(x_1 - x_2)^2 + (y_1 - y_2)^2 + (z_1 - z_2)^2}\right)^3} \cdot (x_2 - x_1) \\
m_1 \cdot \ddot{y}_1 &= \frac{G \cdot m_1 \cdot m_2}{\left(\sqrt{(x_1 - x_2)^2 + (y_1 - y_2)^2 + (z_1 - z_2)^2}\right)^3} \cdot (y_2 - y_1) \\
m_1 \cdot \ddot{z}_1 &= \frac{G \cdot m_1 \cdot m_2}{\left(\sqrt{(x_1 - x_2)^2 + (y_1 - y_2)^2 + (z_1 - z_2)^2}\right)^3} \cdot (z_2 - z_1) \\
m_2 \cdot \ddot{x}_2 &= \frac{G \cdot m_1 \cdot m_2}{\left(\sqrt{(x_1 - x_2)^2 + (y_1 - y_2)^2 + (z_1 - z_2)^2}\right)^3} \cdot (x_1 - x_2) \\
m_2 \cdot \ddot{y}_2 &= \frac{G \cdot m_1 \cdot m_2}{\left(\sqrt{(x_1 - x_2)^2 + (y_1 - y_2)^2 + (z_1 - z_2)^2}\right)^3} \cdot (y_1 - y_2) \\
m_2 \cdot \ddot{z}_2 &= \frac{G \cdot m_1 \cdot m_2}{\left(\sqrt{(x_1 - x_2)^2 + (y_1 - y_2)^2 + (z_1 - z_2)^2}\right)^3} \cdot (z_1 - z_2)
\end{align*}
$$

**b) Transform the following equations into first-order form.**

$$
\begin{align*}
\dot{x_1} &= v_{x_1} \\
\dot{y_1} &= v_{y_1} \\
\dot{z_1} &= v_{z_1} \\
\dot{x_2} &= v_{x_2} \\
\dot{y_2} &= v_{y_2} \\
\dot{z_2} &= v_{z_2} \\
\dot{v}_{x_1} &= \frac{G \cdot m_2}{\left(\sqrt{(x_1 - x_2)^2 + (y_1 - y_2)^2 + (z_1 - z_2)^2}\right)^3} \cdot (x_2 - x_1) \\
\dot{v}_{y_1} &= \frac{G \cdot m_2}{\left(\sqrt{(x_1 - x_2)^2 + (y_1 - y_2)^2 + (z_1 - z_2)^2}\right)^3} \cdot (y_2 - y_1) \\
\dot{v}_{z_1} &= \frac{G \cdot m_2}{\left(\sqrt{(x_1 - x_2)^2 + (y_1 - y_2)^2 + (z_1 - z_2)^2}\right)^3} \cdot (z_2 - z_1) \\
\dot{v}_{x_2} &= \frac{G \cdot m_1}{\left(\sqrt{(x_1 - x_2)^2 + (y_1 - y_2)^2 + (z_1 - z_2)^2}\right)^3} \cdot (x_1 - x_2) \\
\dot{v}_{y_2} &= \frac{G \cdot m_1}{\left(\sqrt{(x_1 - x_2)^2 + (y_1 - y_2)^2 + (z_1 - z_2)^2}\right)^3} \cdot (y_1 - y_2) \\
\dot{v}_{z_2} &= \frac{G \cdot m_1}{\left(\sqrt{(x_1 - x_2)^2 + (y_1 - y_2)^2 + (z_1 - z_2)^2}\right)^3} \cdot (z_1 - z_2)

## 1.2 1D-System Analysis

Consider the differential equation: 

$$
\frac{dx}{dt} = x(1 - x)
$$

**a) Find the fixed points**

$$
\begin{align*}
0 &= \frac{dx}{dt} \\
0 &= x(1 - x) \\
\end{align*}
$$

$$
\begin{align*}
\Rightarrow\ x = 0 \;\text{ or }\; 1 - x = 0 \Rightarrow\
\begin{cases}
x^*_1 = 0 \\
x^*_2 = 1
\end{cases}
\end{align*}
$$

**b) Sketch the phase line**

```python
# Implement differential equation

def dx(x: float) -> float:
    """Compute rate of change of the one-dimensional system."""
    return x * (1 - x)


# Graphical representaion

plt.figure(figsize=(7, 2))

x_values = np.linspace(-1, 2, 10)

# Direction arrows
for x in x_values:
    d = np.sign(dx(x))
    if d > 0:
        plt.arrow(x, 0, 0.1, 0, head_width=0.075, head_length=0.075, fc='black', ec='black', length_includes_head=True)
    elif d < 0:
        plt.arrow(x, 0, -0.1, 0, head_width=0.075, head_length=0.075, fc='black', ec='black', length_includes_head=True)

# Fixed points
fixed_points = np.array([[0, 0], [1, 0]])
plt.scatter(*fixed_points.T, color="red", s=25, zorder=3)

# Plot formatting
plt.axhline(0, color='black')
plt.yticks([])
plt.xlabel("x")
plt.title("Phase Line of 1D System")
plt.xlim(-1, 2)
plt.ylim(-1, 1)
plt.show()
```

### 1.3.1 Eigenvalues I

Given the system:

$$
\begin{align*}
\frac{dx}{dt} = Ax, \qquad A = \begin{pmatrix} 2 & 1 \\ 0 & 3 \end{pmatrix}
\end{align*}

**a) Compute the eigenvalues of A.**

$$
\begin{align*}
0 &= \det(A - \lambda I) \\
0 &= \det\begin{pmatrix} 2 - \lambda & 1 \\ 0 & 3 - \lambda \end{pmatrix} \\
0 &= (2 - \lambda)(3 - \lambda) - 1 \cdot 0 \\
0 &= (2 - \lambda)(3 - \lambda) 
\end{align*}
$$

$$
\begin{align*}
\Rightarrow\ 
\begin{cases}
\lambda_1 = 2 \\
\lambda_2 = 3
\end{cases}
\end{align*}

$$

### 1.4.5 Bacterial Growth with Antibiotics

A bacterial population grows exponentially at rate $r$, but antibiotics are applied that reduce the
population. The killing rate is proportional to both the bacterial population $x(t)$ and the antibiotic
concentration $A(t)$.

**a) Write a differential equation for $x(t)$, the bacterial population.**

$$
\begin{align*}
\dot{x} = r \cdot x - A(t) \cdot x = x (r - A(t))
\end{align*}
$$

**b) Discuss how the model behavior changes for constant, decaying, or pulsed antibiotic concentration $A(t)$**.

For a constant antibiotic concentration, $r$ and $A$ can be treated as constants. The net growth rate is $r - A$.  
- If $r - A > 0$, the population grows exponentially.  
- If $r - A < 0$, the population decays exponentially.  
- If $r - A = 0$, the population remains constant over time.  

For a decaying antibiotic concentration, $A(t)$ decreases with time. Initially, if $A(0) > r$, the population declines. As $A(t)$ becomes smaller, the net rate $r - A(t)$ eventually turns positive, leading to exponential growth. The growth rate approaches $r$ once the antibiotic concentration has effectively vanished.  

For a pulsed antibiotic concentration, $A(t)$ alternates between high and low values. The bacterial population then oscillates: it decreases during high antibiotic phases ($A > r$) and increases during low antibiotic phases ($A < r$). The overall trend depends on the average value of $A(t)$ compared to $r$.
